```python
import pandas as pd
import json
import requests

url = "https://community-open-weather-map.p.rapidapi.com/climate/month"

querystring = {"q":"cairo"}

headers = {
    'x-rapidapi-host': "community-open-weather-map.p.rapidapi.com",
    'x-rapidapi-key': "8b63ef84b8msh5618effa460113bp1bda80jsn7eb4acf88e44"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```

```python
data=json.loads(response.text)
```

```python
df_from_API = pd.DataFrame(json.loads(response.text)['list'])
df_from_API.head(2)
```

```python
df_from_API["temp"]
```

```python
df_from_API = pd.concat([df_from_API, pd.json_normalize(df_from_API['temp'])], axis=1)
df_from_API = df_from_API.rename(columns={'average': 'temp_average', 'average_max': 'temp_average_max', 'average_min': 'temp_average_min',
                                'record_max': 'temp_record_max', 'record_min': 'temp_record_min'})
df_from_API.head(2)
```
